Aby przefiltrować sygnał wykonujemy następujący kod:

```python
import numpy as np
import matplotlib.pyplot as plt
import aseegg as ag
```

```python
czest1 = 5
czest2 = 17
czest3 = 31
czestProbkowania = 250
czas = 3


t = np.linspace(0, czas, czas * czestProbkowania)
sygnal = np.sin(2* np.pi* czest1 *t)+np.sin(2* np.pi*czest2 *t)+np.sin(2* np.pi* czest3 *t)


przefiltrowany = ag.gornoprzepustowy(sygnal, czestProbkowania, 10)
```

```python
plt.subplot(2, 1, 1)
plt.plot(t, sygnal)
plt.xlim([0, 1])
plt.subplot(2, 1, 2)
plt.plot(t, przefiltrowany)
plt.xlim([0, 1])
plt.show()
```

## Jak wygląda ten wcześniej stworzony sygnał w dziedzinie częstotliwości przed i po filtracji? 

```python
ffted_przed_filtracja = ag.FFT(sygnal)
f = np.linspace(0, czestProbkowania, czas * czestProbkowania)

plt.plot(f, ffted_przed_filtracja)
plt.xlim([0,40])
```

```python
ffted_po_filtracji = ag.FFT(przefiltrowany)

plt.plot(f, ffted_po_filtracji)
plt.xlim([0,40])
```
